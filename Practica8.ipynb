Importamos librerías y cargamos el dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Cargamos el dataset tips
df = sns.load_dataset('tips')
```

## Mapa de calor (Heatmap)

Correlación entre variables numéricas

```python
# Seleccionar columnas numéricas
numericas = df[['total_bill', 'tip', 'size']]

# Matriz de correlación
correlacion = numericas.corr()

# Crear mapa de calor
plt.figure(figsize=(8,6))
sns.heatmap(
    correlacion,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Mapa de calor de correlación')
plt.show()
```

## Diagrama de violín

Distribución de propinas por día

```python
plt.figure(figsize=(8,6))
sns.violinplot(
    x='day',
    y='tip',
    data=df,
    palette='Set2'
)

plt.title('Distribución de propinas por día')
plt.xlabel('Día')
plt.ylabel('Propina')
plt.show()
```

## Gráfico de dispersión

Relación entre total de la cuenta y propina

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x='total_bill',
    y='tip',
    hue='day',
    data=df
)

plt.title('Relación entre total de la cuenta y propina')
plt.xlabel('Total de la cuenta')
plt.ylabel('Propina')
plt.legend(title='Día')
plt.show()
```

## Gráfico de barras

Promedio de propinas por género

```python
plt.figure(figsize=(6,5))
sns.barplot(
    x='sex',
    y='tip',
    data=df,
    estimator=np.mean,
    palette='pastel'
)

plt.title('Promedio de propinas por género')
plt.xlabel('Género')
plt.ylabel('Propina promedio')
plt.show()
```
