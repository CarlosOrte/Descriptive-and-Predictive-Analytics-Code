**Creamos un arreglo de 20 elementos aleatorios entre 1 y 100**

```python
import numpy as np

arreglo = np.random.randint(1, 100, 20)

print("Arreglo original:")
print(arreglo)
```

**Buscamos el valor máximo...**

```python
maximo = max(arreglo)
print("\nValor máximo:", maximo)
```

**Luego el valor mínimo...**

```python
minimo = min(arreglo)
print("Valor mínimo:", minimo)
```

**El valor de la media se define de la siguiente manera...**

```python
media = sum(arreglo) / len(arreglo)
print("Media del arreglo:", media)
```

**Finalmente sumamos 10 a cada elemento del arreglo...**

```python
arreglo_mas_10 = arreglo + 10
print("\nArreglo con 10 sumado a cada elemento:")
print(arreglo_mas_10)
```
