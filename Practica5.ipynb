**Crear arreglo de 15 números aleatorios entre 1 y 50**

```python
import numpy as np

arreglo = np.random.randint(1, 51, 15)

print("Arreglo original:")
print(arreglo)
```

**Crear máscara booleana (mayores que 25)**

```python
mascara = arreglo > 25

print("\nMáscara booleana (arreglo > 25):")
print(mascara)
```

**Seleccionar elementos usando la máscara**

```python
mayores_25 = arreglo[mascara]

print("\nElementos mayores que 25:")
print(mayores_25)
```

**Contar cuántos elementos son mayores que 25**

```python
cantidad = np.sum(mascara)

print("\nCantidad de elementos mayores que 25:", cantidad)
```
