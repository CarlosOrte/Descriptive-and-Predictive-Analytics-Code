```python
import pandas as pd

#cargar archivo csv
df = pd.read_csv('/content/drive/MyDrive/nls97.csv', sep=",")
print(df)
#imprimimos los nombres de las columnas
print(df.columns)
```

## Ejercicio 01: Errores Tipográficos y Valores Inconsistentes: maritalstatus

Imprime de la columna 'maritalstatus' los valores únicos.

```python
print(df['maritalstatus'].unique())
```

imprime la frecuencia de cada valor único

```python
print(df['maritalstatus'].value_counts())
```

Cambiamos el valor 'Never-Married' a 'Single'

```python
df['maritalstatus'] = df['maritalstatus'].replace('Never-married', 'Single')

#imprimimos la frecuencia de cada valor único.
print(df['maritalstatus'].value_counts())
```

## Ejercicio 02: Errores Tipográficos y Valores Inconsistentes: highestdegree.

De la columna **'highestdegrees'** imprimimos los **valores únicos** y la **frecuencia de cada valor único**

```python
#Imprimimos los valores únicos
print("valóres únicos: \n", df['highestdegree'].unique())

print()

#Imprimimos la frecuencia de cada valor único
print("frecuencia de cada valor único: \n ", df['highestdegree'].value_counts())
```

Modificamos las categorías para que todas estén en **minúsculas** y empiecen **sin número**

```python

print("valores antes de la modificación:")
print(df['highestdegree'].unique())
print()

#cambiamos cada valor a minúscula
df['highestdegree'] = df['highestdegree'].str.lower()
# quitamos el numero por ejemplo, si dice "2. High School", debería ser 'high school.
df['highestdegree'] = df['highestdegree'].str.replace(r'^\d+\.\s+', '', regex=True)

#imprimimos los valores únicos
print("valores despues:")
print(df['highestdegree'].unique())
```

## Ejercicio 03: Alta Cardinalidad - Frecuencias Bajas

De la columna 'maritalstatus' unimos las categorías Widowed y 'Single' en una categoría más general llamada 'Single/No Partner'. Crea una nueva columna, 'marital_grouped' donde aparezcan las nuevas categorías de maritalstatus.

```python
# imprimimos antes de la transformación
print("valores antes de la modificación:")
print(df['maritalstatus'].unique())
print()

#unimos "Widowed" y "Single" en una categoria general "Single/No Partner"
df['maritalstatus'] = df['maritalstatus'].replace(['Widowed', 'Single'], 'Single/No Partner')

# creamos una columna nueva "marital_grouped"
df['marital_grouped'] = df['maritalstatus']

# Imprimimos los valores únicos de la nueva columna 'marital_grouped'.
print("valores despues de la modificación:")
print(df['marital_grouped'].unique())

print()

# Imprimos la frecuencia de los valores únicos de la nueva columna 'marital_grouped'
print(df['marital_grouped'].value_counts())
```

Imprimos el tipo de datos de la columna gender y la cantidad de bytes
