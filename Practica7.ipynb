## Importar librerías y cargar el dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset Iris
df = sns.load_dataset('iris')
```

## Paso 1. Exploración inicial del dataset

Ver primeras filas

```python
df.head()
```

Información general

```python
df.info()
```

Estadísticas básicas

```python
df.describe()
```

## a) Gráfico de barras

Promedio del largo y ancho del pétalo por especie

```python
# Calcular promedios por especie
petal_means = df.groupby('species')[['petal_length', 'petal_width']].mean()

# Crear gráfico de barras
petal_means.plot(kind='bar', figsize=(8,6))

plt.title('Promedio del largo y ancho del pétalo por especie')
plt.xlabel('Especie')
plt.ylabel('Promedio (cm)')
plt.legend(['Petal Length', 'Petal Width'])
plt.show()
```

## b) Histograma

Distribución del largo del sépalo

```python
plt.figure(figsize=(8,6))
plt.hist(df['sepal_length'], bins=20)

plt.title('Distribución del largo del sépalo')
plt.xlabel('Largo del sépalo (cm)')
plt.ylabel('Frecuencia')
plt.show()
```

## c) Gráfico de dispersión (scatter plot)

Largo del pétalo vs ancho del pétalo

```python
plt.figure(figsize=(8,6))

species = df['species'].unique()

for s in species:
    subset = df[df['species'] == s]
    plt.scatter(subset['petal_length'], subset['petal_width'], label=s)

plt.title('Relación entre largo y ancho del pétalo')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend()
plt.show()
```

## d) Gráfico de cajas (box plot)

Distribución de sépalos por especie

```python
plt.figure(figsize=(10,6))

df.boxplot(column=['sepal_length', 'sepal_width'], by='species')

plt.title('Distribución de medidas del sépalo por especie')
plt.suptitle('')
plt.xlabel('Especie')
plt.ylabel('Medida (cm)')
plt.show()
```

## e) Gráfico de líneas

Tendencia del largo del pétalo a lo largo del dataset

```python
plt.figure(figsize=(8,6))
plt.plot(df.index, df['petal_length'])

plt.title('Tendencia del largo del pétalo')
plt.xlabel('Índice de observación')
plt.ylabel('Petal Length (cm)')
plt.show()
```
