##Ejercicio 02 Duplicidad Parcial Dos columnaa: Contar y Filtrar los registros duplicados en Pulse y MaxPulse

```python
# =========================================================
# Duplicidad parcial en dos columnas: Pulse y MaxPulse
# =========================================================

df2 = pd.read_csv(archivo)

print("Primeras filas:")
print(df2.head())

# Contar registros duplicados usando Pulse y MaxPulse
duplicados_pulse_maxpulse = df2.duplicated(subset=["Pulse", "MaxPulse"]).sum()

print("\nCantidad de registros duplicados en Pulse y MaxPulse:")
print(duplicados_pulse_maxpulse)

# value_counts de combinaciones Pulse y MaxPulse
print("\nCantidad de registros por cada combinación única de Pulse y MaxPulse:")
print(df2[["Pulse", "MaxPulse"]].value_counts())

# Filtrar registros y conservar la última ocurrencia
df2_keep_last = df2.drop_duplicates(subset=["Pulse", "MaxPulse"], keep="last")

print("\nDataFrame sin duplicados en Pulse y MaxPulse (keep='last'):")
print(df2_keep_last)

print("\nDimensiones originales:", df2.shape)
print("Dimensiones con keep='last':", df2_keep_last.shape)
```

##Ejercicio 03 Duplicidad Total o Exacta: Contar y Filtrar los registros duplicado

```python
# =========================================================
# Duplicidad total o exacta
# =========================================================

df3 = pd.read_csv(archivo)

print("Primeras filas:")
print(df3.head())

# Identificar duplicados exactos en todas las columnas
duplicados_exactos = df3.duplicated()

# Contar cantidad de duplicados exactos usando sum()
cantidad_duplicados_exactos = duplicados_exactos.sum()

print("\nCantidad de registros duplicados exactos:")
print(cantidad_duplicados_exactos)

# Filtrar registros conservando la última ocurrencia
df3_keep_last = df3.drop_duplicates(keep="last")

print("\nDataFrame sin duplicados exactos (keep='last'):")
print(df3_keep_last)

print("\nDimensiones originales:", df3.shape)
print("Dimensiones sin duplicados exactos:", df3_keep_last.shape)
```

##Ejercicio 04: Cálculo de la media con duplicados y sin duplicados

```python
# =========================================================
# Media con duplicados y sin duplicados
# =========================================================

df4 = pd.read_csv(archivo)

print("Primeras filas:")
print(df4.head())

# Media original con duplicados
media_original_pulse = df4["Pulse"].mean()
media_original_maxpulse = df4["MaxPulse"].mean()

print("\nMedia original de Pulse:")
print(media_original_pulse)

print("\nMedia original de MaxPulse:")
print(media_original_maxpulse)

# Eliminar duplicados parciales en Pulse y MaxPulse conservando las últimas ocurrencias
df4_sin_duplicados = df4.drop_duplicates(subset=["Pulse", "MaxPulse"], keep="last")

print("\nDataFrame sin duplicados en Pulse y MaxPulse (keep='last'):")
print(df4_sin_duplicados)

# Media sin duplicados
media_sin_dup_pulse = df4_sin_duplicados["Pulse"].mean()
media_sin_dup_maxpulse = df4_sin_duplicados["MaxPulse"].mean()

print("\nMedia de Pulse sin duplicados:")
print(media_sin_dup_pulse)

print("\nMedia de MaxPulse sin duplicados:")
print(media_sin_dup_maxpulse)
```
