```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1) Cargar dataset penguins
df = sns.load_dataset("penguins")

# 2) Crear figura 2x2
fig, ax = plt.subplots(2, 2, figsize=(12, 8))

# (0,0) Histograma de longitudes de aletas
sns.histplot(data=df, x="flipper_length_mm", ax=ax[0,0], kde=True, color="skyblue")
ax[0,0].set_title("Histograma: longitud de aletas")

# (0,1) Boxplot masa corporal por especie
sns.boxplot(data=df, x="species", y="body_mass_g", ax=ax[0,1], palette="Set2")
ax[0,1].set_title("Boxplot: masa corporal por especie")
ax[0,1].set_xlabel("Especie")
ax[0,1].set_ylabel("Masa corporal (g)")

# (1,0) Dispersión masa corporal vs largo de aletas
sns.scatterplot(data=df, x="flipper_length_mm", y="body_mass_g", hue="species", ax=ax[1,0])
ax[1,0].set_title("Dispersión: masa corporal vs largo de aletas")
ax[1,0].set_xlabel("Longitud de aletas (mm)")
ax[1,0].set_ylabel("Masa corporal (g)")

# (1,1) Violin plot masa corporal según especie
sns.violinplot(data=df, x="species", y="body_mass_g", ax=ax[1,1], palette="Pastel1")
ax[1,1].set_title("Violin plot: masa corporal por especie")
ax[1,1].set_xlabel("Especie")
ax[1,1].set_ylabel("Masa corporal (g)")

# Ajustar diseño
plt.tight_layout()
plt.show()
```
