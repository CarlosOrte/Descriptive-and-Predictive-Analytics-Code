## Práctica 12: Carga de Datos

```python
# ===============================
# CARGA DEL DATASET
# ===============================

import pandas as pd

# Cargar archivo CSV
df = pd.read_csv('/content/drive/MyDrive/miaad-nyc-r-s.csv', skipfooter=1, engine='python')
df.tail(3)
```

```python
# Numero de fila y columnas en el dataframe
print("Número de filas y columnas:", df.shape)
```

## Práctica 13: Cambiar nombre de columna

```python
# imprimimos el nombre de las columnas
print(df.columns)
```

```python
# cambiamos el nombre de la primer columna a INDEX MIAAD
df.rename(columns={df.columns[0]: "INDEX MIAAD"}, inplace=True)
```

```python
print("Primera columna actual:", df.columns[0])
```

## Práctica 14: Normalización de los nombres de la columna.

```python
# convertimos las columnas a minúsculas y remplazamos espacios por "_"
df.columns = df.columns.str.lower().str.replace(" ", "_")

# imprimimos el nombre de las columnas
print(df.columns)
```

## Práctica 15: Errores en los Tipos de Datos

```python
# imprimimos los tipos de datos de cada columna
print(df.dtypes)
```

```python
# verificamos que los tipos de datos hay en cada columna
df.head()
```

```python
# cambiamos las columnas sale_price, land_square_feet, gross_square_feet a punto flotante
# Primero, reemplazamos ' - ' con NaN y luego convertimos a tipo numérico, forzando los errores a NaN
df['sale_price'] = pd.to_numeric(df['sale_price'].replace(' -  ', pd.NA), errors='coerce')
df['land_square_feet'] = pd.to_numeric(df['land_square_feet'].replace(' -  ', pd.NA), errors='coerce')
df['gross_square_feet'] = pd.to_numeric(df['gross_square_feet'].replace(' -  ', pd.NA), errors='coerce')

# imprimimos los tipos de datos de cada columna (de nuevo)
print(df.dtypes)
```

## Práctica 16: Genera el diccionario de datos del dataframe preprocesado. Imprime el contenido de dicho diccionario

```python
#Crea un nuevo DataFrame llamado diccionario_datos que almacena
# la descripción de los nombres de las columnas y sus tipos de datos.
#[str(df[col].dtype) for col in df.columns]:
#Itera sobre cada columna (col) de df.
#Obtiene su tipo de dato con df[col].dtype.
#Convierte el tipo de dato a una cadena (str) para que pueda almacenarse en el nuevo DataFrame.
diccionario_datos = pd.DataFrame({"Nombre de columna": df.columns,
                                  "Tipo de dato": [str(df[col].dtype) for col in df.columns]})

#Imprimimos el diccionario
print(diccionario_datos)

#guardamos el diccionario
diccionario_datos.to_csv("/content/drive/MyDrive/diccionario_datos.csv", index=False, encoding="utf-8")
```
