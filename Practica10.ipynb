# Importar librerías


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1) DataFrame de ventas
np.random.seed(42)

meses = ["Ene","Feb","Mar","Abr","May","Jun","Jul","Ago","Sep","Oct","Nov","Dic"]
df = pd.DataFrame({
    "Mes": meses,
    "Ventas": np.random.randint(2000, 5001, size=12),
    "Gastos": np.random.randint(1000, 4001, size=12)
})

# 2) GridSpec para subplots de diferentes tamaños
fig = plt.figure(figsize=(12, 8))
gs = fig.add_gridspec(2, 3)

# Primer gráfico: barra, ocupando dos columnas (arriba izquierda)
ax1 = fig.add_subplot(gs[0, 0:2])
ax1.bar(df["Mes"], df["Ventas"])
ax1.set_title("Ventas por mes (Barras)")
ax1.set_xlabel("Mes")
ax1.set_ylabel("Ventas")

# Segundo gráfico: línea, una columna (arriba derecha)
ax2 = fig.add_subplot(gs[0, 2])
ax2.plot(df["Mes"], df["Gastos"], marker="o")
ax2.set_title("Gastos por mes (Línea)")
ax2.set_xlabel("Mes")
ax2.set_ylabel("Gastos")

# Tercer gráfico: dispersión, dos filas (columna derecha completa en filas restantes)
ax3 = fig.add_subplot(gs[1, :])
ax3.scatter(df["Ventas"], df["Gastos"])
ax3.set_title("Ventas vs Gastos (Dispersión)")
ax3.set_xlabel("Ventas")
ax3.set_ylabel("Gastos")

# 3) Ajustar espacios
plt.tight_layout()
plt.show()
```
