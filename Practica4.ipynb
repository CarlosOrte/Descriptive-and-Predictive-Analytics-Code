Creamos dos matrices 3x3 con numeros aleatorios entre 1 y **10**

```python
import numpy as np

# Crear dos matrices 3x3 con valores aleatorios entre 1 y 10
matriz1 = np.random.randint(1, 11, (3, 3))
matriz2 = np.random.randint(1, 11, (3, 3))

print("Matriz 1:")
print(matriz1)

print("\nMatriz 2:")
print(matriz2)
```

**1. Suma de matrices**

```python
suma = matriz1 + matriz2
print("\nSuma de las matrices:")
print(suma)
```

**2. Resta de la segunda matriz a la primera**

```python
resta = matriz1 - matriz2
print("\nResta (Matriz 1 - Matriz 2):")
print(resta)
```

**3. Multiplicar la primera matriz por 2**

```python
multiplicacion_escalar = matriz1 * 2
print("\nMatriz 1 multiplicada por 2:")
print(multiplicacion_escalar)
```

**4. Multiplicación matricial**

```python
multiplicacion_matricial = np.dot(matriz1, matriz2)
print("\nMultiplicación matricial (Matriz 1 x Matriz 2):")
print(multiplicacion_matricial)
```
