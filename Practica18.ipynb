## Ejercicio 01: Tipo de Datos Correcto

```python
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv('/content/drive/MyDrive/Airdata.csv')

# Mostrar tipos de datos antes de convertir
print("Tipos de datos antes de la conversión:\n")
print(df.dtypes)

# Convertir la columna DateTime a tipo datetime si no está correcta
df['DateTime'] = pd.to_datetime(df['DateTime'], errors='coerce')

# Mostrar tipos de datos después de convertir
print("\nTipos de datos después de la conversión:\n")
print(df.dtypes)

# Imprimir las primeras 3 líneas
print("\nPrimeras 3 filas del dataframe:\n")
print(df.head(3))
```

## Ejercicio 02: Extraer componentes de fecha y tiempo

```python
# Crear nuevas columnas a partir de DateTime
df['Year'] = df['DateTime'].dt.year
df['Month'] = df['DateTime'].dt.month
df['Day'] = df['DateTime'].dt.day
df['Hour'] = df['DateTime'].dt.hour

# Imprimir las primeras 3 líneas
print("Primeras 3 filas con nuevas columnas:\n")
print(df.head(3))
```

## Ejercicio 03: Aritmética - Suma de Intervalos (timedelta)

```python
# Sumar 10 días, 7 horas y 15 minutos a DateTime
df['DateTime_plus'] = df['DateTime'] + pd.Timedelta(days=10, hours=7, minutes=15)

# Imprimir solo las columnas pedidas
print("Columnas DateTime y DateTime_plus:\n")
print(df[['DateTime', 'DateTime_plus']].head(10))
```

## Ejercicio 04: Timestamp

```python
# Convertir DateTime a timestamp Unix (segundos desde 1970-01-01)
df['UnixTimestamp'] = df['DateTime'].astype('int64') // 10**9

# Imprimir primeras 3 filas de ambas columnas
print("Primeras 3 filas con DateTime y UnixTimestamp:\n")
print(df[['DateTime', 'UnixTimestamp']].head(3))
```

## Ejercicio 05: Filtrado de Fechas

```python
# Filtrar fechas mayores a 2020-11-01
filtro = df[df['DateTime'] > '2020-11-01']

# Imprimir resultado
print("Fechas mayores a 2020-11-01:\n")
print(filtro)
```
