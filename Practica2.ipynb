Lectura del **Archivo**

```python
# Abrir y leer el archivo
with open('/content/drive/MyDrive/ClassFiles/informe_acciones.txt', 'r', encoding='latin-1') as archivo:
     # Recorrer cada línea del archivo
    for linea in archivo:
           # Buscar el patrón "Valor actual: $numero"
        resultado = re.search(r'Valor actual:\s*\$([0-9]+\.[0-9]+)', linea)

        # Si se encuentra el valor
        if resultado:
            valor = resultado.group(1)
            print(valor)
```
